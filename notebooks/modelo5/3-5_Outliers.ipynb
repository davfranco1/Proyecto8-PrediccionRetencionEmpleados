- Vemos que, nos hemos quedado sin outliers en el DF filtrado.

```python
sp.detectar_outliers(df_filtrado)
```

- En total, hemos identificado sólamente 15 outliers, de un total de unos 1600 registros. Por ese motivo, en este segundo modelo, los eliminaremos del dataset.

```python
df = df.drop(df_filtrado.index)
```
