- Importamos el dataframe generado en el notebook anterior.

```python
df = pd.read_pickle('../../datos/pickles_transformados/modelo3/df_sinnulos.pkl')
df.head()
```

- Para saber si las variables tienen orden y el tipo de encoding que debemos aplicar a cada una, aplicaremos el método del chi cuadrado.

```python
v_ord, v_no_ord = se.detectar_orden_cat(df, lista_cat=df.select_dtypes(include = ["object", "category"]), var_respuesta="Attrition")
```

- Dado que, los modelos predictivos que utilizamos no aceptan datos de tipo string, y tenemos una variable respuesta binaria, mapearemos sus valores de "Yes" y "No" a 1 y 0.

```python
df["Attrition"] = df["Attrition"].map({"Yes":1,"No":0})
```

- Aunque sabemos que, cuando las variables **no** tienen orden (p-valor >0,05), debemos usar el One-Hot Encoder, en este caso, por la cantidad de variables que tenemos sin orden (8) y la gran cantidad de categorías en ellas, aumentaríamos significativamente la dimensionalidad del dataset, lo cual no es práctico. Por ello, la llevaremos al mismo método que usaremos para las ordenadas.

- En el caso de las 8 variables ordenadas, usaremos el Target Encoder. Usaremos un diccionario del que "beberá" la clase "Encoding", para saber qué modelo aplicar.

```python
diccionario_encoding = {
    "onehot": [],
    "dummies": [],
    "ordinal": {},
    "label": [],
    "frequency": [],
    "target": ['Gender', 'Education', 'JobLevel', 'StockOptionLevel', 'PerformanceRating', 'BusinessTravel', 'Department', 'EducationField', 'JobRole', 'MaritalStatus', 'TrainingTimesLastYear', 'EnvironmentSatisfaction', 'JobSatisfaction', 'WorkLifeBalance', 'JobInvolvement'],
    "catboost": []
}
```

- Instanciamos la clase y aplicamos el encoder elegido.

- Usaremos el Target Encoder dado que, convierte las categorías en valores numéricos basados en el promedio de la variable objetivo ("Attrition") para cada categoría. Como mencionamos antes, el Target Encoding puede ser más eficiente que One-Hot Encoding, que generaría demasiadas columnas. Además, si las categorías están directamente relacionadas con la probabilidad del target, el Target Encoder puede capturar esta información.

```python
encoding = se.Encoding(dataframe=df, diccionario_encoding=diccionario_encoding, variable_respuesta="Attrition")
```

```python
ta_enc, encoder = encoding.target_encoding()
ta_enc.head()
```
