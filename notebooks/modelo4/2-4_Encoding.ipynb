- Dado que, los modelos predictivos que utilizamos no aceptan datos de tipo string, y tenemos una variable respuesta binaria, mapearemos sus valores de "Yes" y "No" a 1 y 0.

- Aunque sabemos que, cuando las variables **no** tienen orden (p-valor >0,05), debemos usar el One-Hot Encoder, en este caso, por la cantidad de variables que tenemos sin orden (8) y la gran cantidad de categorías en ellas, aumentaríamos significativamente la dimensionalidad del dataset, lo cual no es práctico. Por ello, la llevaremos al mismo método que usaremos para las ordenadas.

- En el caso de las 8 variables ordenadas, usaremos el Catboost Encoder. Usaremos un diccionario del que "beberá" la clase "Encoding", para saber qué modelo aplicar.

```python
diccionario_encoding = {
    "onehot": [],
    "dummies": [],
    "ordinal": {},
    "label": [],
    "frequency": [],
    "target": [],
    "catboost": ['Gender', 'Education', 'Gender', 'JobLevel', 'StockOptionLevel', 'PerformanceRating', 'Attrition', 'BusinessTravel', 'Department', 'EducationField', 'JobRole', 'MaritalStatus', 'TrainingTimesLastYear', 'EnvironmentSatisfaction', 'JobSatisfaction', 'WorkLifeBalance', 'JobInvolvement']
}
```

- Instanciamos la clase y aplicamos el encoder elegido.

- Usaremos el CatBoost Encoder, que es una técnica de codificación similar al Target Encoder, pero mejora el manejo del overfitting. Está diseñado específicamente para problemas supervisados y utiliza un enfoque basado en medias condicionadas a la variable objetivo (como el Target). Para cada fila, excluye su propio target al calcular la media de su categoría, reduciendo la “fuga” de información entre entrenamiento y prueba, reduciendo el overfitting. Similar al Target, es ideal para manejar variables con muchas categorías únicas.

```python
encoding = se.Encoding(dataframe=df, diccionario_encoding=diccionario_encoding, variable_respuesta="Attrition")
```

```python
cb_enc, encoder = encoding.catboost_encoding()
cb_enc.head()
```
